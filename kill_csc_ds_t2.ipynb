```python
df  = pd.read_csv('./processes2.csv')
df
```

# Describing Data

```python
print('\n Dataset Summary : \n')
pd.concat([
     pd.DataFrame(df.count()).T.rename(index={0: 'count'}),
    pd.DataFrame(df.nunique()).T.rename(index={0: 'number of unique'}),
    pd.DataFrame(df.dtypes).T.rename(index={0: 'dtype'}),
    pd.DataFrame(df.isna().sum()).T.rename(index={0: 'null count'}),
    df.describe().drop('count')
]).T
```

## Dropping an Unnecessary columns

```python
df.drop(columns = ['Mileage Unit'], inplace = True)
df.head()
```

```python
print('Categorical values of name : ', df.name.unique())
print('Categorical values of fuel : ', df.fuel.unique())
print('Categorical values of seller_type : ', df.seller_type.unique())
print('Categorical values of owner : ', df.owner.unique())
print('Categorical values of transmission : ', df.transmission.unique())
```

### Converting Categorical Data to Numerical Data

```python
df_tensor = df.copy()
name = {'Maruti':11 ,'Hyundai':10 , 'Toyota' :9,'Ford':8, 'Mahindra':6, 'Tata' :5, 'Volkswagen':4,'Renault' :3,'Chevrolet' :2, 'Honda' :1}
fuel = {'Diesel' :4, 'Petrol':3, 'CNG' :2,'LPG':1}
seller_type = {'Individual':3, 'Dealer':2, 'Trustmark Dealer':1}
owner = {'First Owner':5, 'Second Owner':4 ,'Third Owner':3, 'Fourth & Above Owner':2,'Test Drive Car':1}
transmission = {'Manual' :2, 'Automatic':1}

df_tensor['name'] = df_tensor['name'].map(name)
df_tensor['fuel'] = df_tensor['fuel'].map(fuel)
df_tensor['seller_type'] = df_tensor['seller_type'].map(seller_type)
df_tensor['owner'] = df_tensor['owner'].map(owner)
df_tensor['transmission'] = df_tensor['transmission'].map(transmission)
```

```python
df_tensor
```

```python
X = df_tensor.drop(columns = ['seller_type','Unnamed: 0', 'max_power (in bph)','owner','name' ])
```

```python
X = df_tensor.drop('selling_price', axis = 'columns')
X
```

```python
y = df_tensor.selling_price
y
```

### sklearn.model_selection

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.2)
```

```python
from sklearn.linear_model import LinearRegression
classifier = LinearRegression()
```

```python
classifier.fit(X_train, y_train)
```

# Accuracy Score Of my Model

```python
classifier.score(X_test, y_test)
```
